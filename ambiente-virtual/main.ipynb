```python
#importando as bibliotecas e criando a conexão

import psycopg2
from psycopg2 import sql

conn = psycopg2.connect(
    host='localhost',
    port=5432,
    dbname='postgres',
    user='postgres',
    password='root'
)

conn.autocommit = True 
```

```python
db_name = 'db_senai'

create_db_query = sql.SQL('CREATE DATABASE {}').format(sql.Identifier(db_name))

cur = conn.cursor();

cur.execute(create_db_query);

print(f'Bando de dados {db_name} criado com sucesso')
```

```python
create_table_query = '''
    CREATE TABLE Aluno(
        id bigint primary key,
        nome_aluno varchar(80) not null,
        data_nasc date not null
    )
'''

cur.execute(create_table_query)
conn.commit()

print("Tabela criada com sucesso")
```

```python
insert_query = '''
    INSERT INTO Aluno(id, nome_aluno, data_nasc) 
    VALUES(1, 'Vicente', '2007-05-11'),
    (2, 'Maria', '2006-07-23')
    '''

cur.execute(insert_query)
conn.commit()

print("Dados inseridos")
```

```python
cur.execute('SELECT * FROM Aluno')
rows = cur.fetchall()

for row in rows:
    print(row)
```

```python
update_query = '''
    UPDATE Aluno set nome_aluno = %s WHERE nome_aluno = %s'''
    
cur.execute(update_query, ('Pedro', 'Vicente'))
conn.commit()

cur.execute('SELECT * FROM Aluno')
rows = cur.fetchall()

for row in rows:
    print(row)
```

```python
cur.execute('DELETE FROM Aluno WHERE nome_aluno = %s', ('Maria',))
conn.commit()

cur.execute('SELECT * FROM Aluno')
rows = cur.fetchall()

for row in rows:
    print(row)
```
